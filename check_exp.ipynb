```python
import numpy as np
import matplotlib.pyplot as plt

# Time variable
t = np.linspace(0, 120,1000)  # Time from 0 to 10 seconds

# Define the pitch and sway functions for each trial
trials = {
    16: {
        'pitch': lambda t: 6 * np.sin(2 * np.pi * t / 4),
        'sway': lambda t: 0.035 * np.cos(2 * np.pi * t / 3),
    },
    17: {
        'pitch': lambda t: 4 * np.cos(2 * np.pi * t / 2),
        'sway': lambda t: 0.03 * np.sin(2 * np.pi * t / 3),
    },
    18: {
        'pitch': lambda t: 7 * np.sin(2 * np.pi * t / 4),
        'sway': lambda t: 0.025 * np.cos(2 * np.pi * t / 2),
    },
    19: {
        'pitch': lambda t: 5 * np.cos(2 * np.pi * t / 7),
        'sway': lambda t: 0.035 * np.sin(2 * np.pi * t / 6),
    },
    20: {
        'pitch': lambda t: 6 * np.sin(2 * np.pi * t / 6),
        'sway': lambda t: 0.03 * np.cos(2 * np.pi * t / 5),
    },
    21: {
        'pitch': lambda t: 3 * np.cos(2 * np.pi * t / 3),
        'sway': lambda t: 0.04 * np.sin(2 * np.pi * t / 2),
    },
    22: {
        'pitch': lambda t: 8 * np.sin(2 * np.pi * t / 2),
        'sway': lambda t: 0.025 * np.cos(2 * np.pi * t / 3),
    },
    23: {
        'pitch': lambda t: 4 * np.sin(2 * np.pi * t / 7),
        'sway': lambda t: 0.03 * np.cos(2 * np.pi * t / 4),
    },
    24: {
        'pitch': lambda t: 5 * np.sin(2 * np.pi * t / 3),
        'sway': lambda t: 0.035 * np.cos(2 * np.pi * t / 5),
    },
    25: {
        'pitch': lambda t: 7 * np.cos(2 * np.pi * t / 6),
        'sway': lambda t: 0.04 * np.sin(2 * np.pi * t / 2),
    },
    26: {
        'pitch': lambda t: 5 * np.sin(2 * np.pi * t / 3) + 3 * np.exp(-0.05 * t) * np.cos(2 * np.pi * t / 5),
        'sway': lambda t: 0.03 * np.sin(2 * np.pi * t / 4) + 0.02 * np.exp(-0.03 * t) * np.cos(2 * np.pi * t / 6),
    },
    27: {
        'pitch': lambda t: 6 * np.exp(-0.02 * t) * np.sin(2 * np.pi * t / 4) + 2 * np.sin(2 * np.pi * t / 7),
        'sway': lambda t: 0.02 * np.cos(2 * np.pi * t / 5) + 0.015 * np.exp(-0.03 * t) * np.sin(2 * np.pi * t / 7.3),
    },
    28: {
        'pitch': lambda t: 4 * np.cos(2 * np.pi *0.005 *t**2 / 5) + 3 * np.exp(-0.02 * t) * np.sin(2 * np.pi *0.005*t**1.5 / 3),
        'sway': lambda t: 0.025 * np.exp(-0.02 * t) * np.sin(2 * np.pi * 0.01*t**2 / 3 ) + 0.03 * np.cos(2 * np.pi * 0.01*t**1.5 / 4),
    },
    29: {
        'pitch': lambda t: 7 * np.sin(2 * np.pi * t / 4) + 2 * np.exp(-0.02 * t) * np.cos(2 * np.pi * 0.01*t**2 / 3),
        'sway': lambda t: 0.025 * np.cos(2 * np.pi * t / 5) + 0.015 * np.exp(-0.02 * t) * np.sin(2 * np.pi * 0.02*t**2 / 3),
    },
    30: {
        'pitch': lambda t: 3 * np.exp(-0.03 * t) * np.cos(2 * np.pi * 0.02*t**2 / 3) + 5 * np.sin(2 * np.pi * t / 5),
        'sway': lambda t: 0.03 * np.sin(2 * np.pi * 0.02*t**2 / 7) + 0.025 * np.exp(-0.03 * t) * np.cos(2 * np.pi * t / 4),
    },
}

# Create separate plots for each trial
for trial, functions in trials.items():
    # Create a new figure for each trial
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    
    # Plot pitch
    ax1.plot(t, functions['pitch'](t))
    ax1.set_title(f'Trial {trial} - Pitch')
    ax1.set_xlabel('Time (s)')
    ax1.set_ylabel('Pitch (deg)')
    
    # Add dashed red lines at +10 and -10
    ax1.axhline(y=10, color='r', linestyle='--')
    ax1.axhline(y=-10, color='r', linestyle='--')

    # Plot sway
    ax2.plot(t, functions['sway'](t))
    ax2.set_title(f'Trial {trial} - Sway')
    ax2.set_xlabel('Time (s)')
    ax2.set_ylabel('Sway (m)')
    
    # Set y-axis limits for sway plot
    ax2.axhline(y=0.05, color='r', linestyle='--')
    ax2.axhline(y=-0.05, color='r', linestyle='--')

    # Add spacing between subplots
    plt.tight_layout()
    
    # Show the plot
    plt.show()
```
